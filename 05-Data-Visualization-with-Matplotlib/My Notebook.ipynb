```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
plt.xlabel('Hehe its x')
plt.ylabel('Hehe its y')
plt.title('Hehe its the title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r')

plt.subplot(1,3,2)
plt.plot(y,x,'b')

plt.subplot(1,3,3)
plt.plot(y,x,'r')
```

Object Oriented

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.set_xlabel('Set X Label') 
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
```
